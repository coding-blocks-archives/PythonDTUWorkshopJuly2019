```
z = 6*np.eye((5)) # Gives you an identity matrix
print(z)
```

```
n = 5
#Update the upper Triangle
for i in range(n):
    for j  in range(n):
        if j>i:
            z[i][j] = 4
            
print(z)
#Updating subsection
z[2:4, 2:4] = 8

#Updating a particular element
z[0][1] = 14
print(z)
```

```
a = np.array([[1,2],[3,4]],dtype='int32')
det  = int(np.linalg.det(a))
```

```
# Inverse of the Matrix
a_inv = np.linalg.pinv(a)
print(a_inv)
```

```
np.round(np.dot(a,a_inv))
```

```
a = np.array([[1,2,3]])

print(a)
print(a.T)
```

```
print(m)
```

```
c = m + 2
print(c)
```

```
combined = np.vstack((m,c))
print(combined)
```
